```python
import pandas as pd

df = pd.read_csv('games.csv')
```

```python
df = df.drop('Other_Sales', axis=1)
df = df.drop('Global_Sales', axis=1)
df = df.drop('Developer', axis=1)
```

```python
df = df.dropna(axis = 0)
df = df.loc[ df['NA_Sales'] > 1]
df = df.loc[ df['EU_Sales'] > 1]
```

```python
names = df.Name
df = df.drop('Name', axis=1)
```

```python
X = df.iloc[:, [0, 1, 2, 3, 7, 8, 9, 10, 11]].values
```

```python
sales_na = df.iloc[:, 4].values
sales_eu = df.iloc[:, 5].values
sales_jp = df.iloc[:, 6].values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = [0, 2, 3, 8]

column_transformer = ColumnTransformer(
    [('onehot', OneHotEncoder(), categorical_features)],
    remainder='passthrough'
)

X = column_transformer.fit_transform(X).toarray()
```

```python
from keras.layers import Input, Dense, Dropout, Activation
from keras.models import Model

entry_layer = Input(shape=(61))
ocult_layer = Dense(units=32, activation='sigmoid')(entry_layer)
ocult_layer_2 = Dense(units=32, activation='sigmoid')(ocult_layer)
out_layer = Dense(units = 1, activation = 'linear')(ocult_layer_2)
out_layer_2 = Dense(units = 1, activation='linear')(ocult_layer_2)
out_layer_3 = Dense(units = 1, activation='linear')(ocult_layer_2)

model = Model(inputs=entry_layer, outputs=[out_layer, out_layer_2, out_layer_3])
model.compile(optimizer='adam', loss='mse')
model.fit(X, [sales_na, sales_eu, sales_jp], epochs=5000, batch_size=100)
```

```python
prevision_na, prevision_eu, prevision_jp = model.predict(X)
```
