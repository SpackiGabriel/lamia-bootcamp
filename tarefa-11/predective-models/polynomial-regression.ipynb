```python
import pandas as pd
import sklearn.model_selection as ms
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataset = pd.read_csv("position-salaries.csv")

x = dataset.iloc[:, 1].values.reshape(-1, 1)
y = dataset.iloc[:, -1]
```

```python
plt.scatter(x, y, color='red')
```

```python
linear_regressor = lm.LinearRegression()
linear_regressor.fit(x, y)
```

```python
plt.scatter(x, y, color='red')
plt.plot(x,  linear_regressor.predict(x), color='blue', scaley=False)
plt.title('Linear simples')
plt.show()
```

```python
linear_regressor.predict([[7.5]])
```

```python
poly_regresssor = PolynomialFeatures(degree=2)
x_poly = poly_regresssor.fit_transform(x)

linear_regressor = lm.LinearRegression()
linear_regressor.fit(x_poly, y)
```

```python
plt.scatter(x, y, color='red')
plt.plot(x,  linear_regressor.predict(x_poly), color='blue', scaley=False)
plt.title('Polinomial')
plt.show()
```

```python
linear_regressor.predict( poly_regresssor.fit_transform([[7.5]]) )
```

```python
poly_regresssor = PolynomialFeatures(degree=4)
x_poly = poly_regresssor.fit_transform(x)

linear_regressor = lm.LinearRegression()
linear_regressor.fit(x_poly, y)
```
